```python
%%time

import matplotlib.pyplot as plt

import torch
import numpy as np
from PIL import Image

from diffusers import ControlNetModel, StableDiffusionXLControlNetPipeline, AutoencoderKL
from diffusers.utils import load_image

controlnet = ControlNetModel.from_pretrained(
    "diffusers/controlnet-depth-sdxl-1.0",
    variant="fp16",
    use_safetensors=True,
    torch_dtype=torch.float16).to("cuda")
vae = AutoencoderKL.from_pretrained("madebyollin/sdxl-vae-fp16-fix", torch_dtype=torch.float16).to("cuda")

pipe = StableDiffusionXLControlNetPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    controlnet=controlnet,
    vae=vae,
    variant="fp16",
    use_safetensors=True,
    torch_dtype=torch.float16).to("cuda")
pipe.enable_model_cpu_offload()
```

```python
%%time

image1 = load_image('./Image generation/1.jpg')
prompt = ["New modern style of livingroom with window, books, chairs, lamp, warm color palette, detailed, 8k"]
negative_prompt = ["low quality, bad quality, sketches"]

with torch.no_grad():
    images1 = pipe(prompt, negative_prompt=negative_prompt, image=image1, guidance_scale=13, num_images_per_prompt=1, num_inference_steps=30, controlnet_conditioning_scale=0.5).images
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 12))

axes[0].imshow(image1,cmap='grey')
axes[0].axis('off')

axes[1].imshow(images1[0])
axes[1].axis('off')

plt.subplots_adjust(top=0.95)
plt.show()
```
